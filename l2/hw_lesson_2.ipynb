```
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
```

Загружаем данные из прошлого урока

```
df = pd.read_pickle("/content/result.pkl")
```

```
df.head()
```

Преобразовываем листы в строки

```
df['tweet_stemmed'] = df['tweet_stemmed'].apply(lambda x: " ".join(x))
df['tweet_lemmatized'] = df['tweet_lemmatized'].apply(lambda x: " ".join(x))
```

Смотрим на CountVectorizer

```
c_bow_stemmed = CountVectorizer(max_df=0.9, max_features=1000, stop_words='english')
c_bow_stemmed_X = c_bow_stemmed.fit_transform(df['tweet_stemmed'])
```

```
c_stemmed_df = pd.DataFrame(c_bow_stemmed_X.toarray(), columns=c_bow_stemmed.get_feature_names())
```

```
c_stemmed_df.head()
```

Смотрим на TfidfVectorizer

```
tfid_bow_lemmatized = TfidfVectorizer(max_df=0.9, max_features=1000, stop_words='english')
tfid_bow_lemmatized_X = tfid_bow_lemmatized.fit_transform(df['tweet_lemmatized'])
```

```
tfid_stemmed_df = pd.DataFrame(tfid_bow_lemmatized_X.toarray(), columns=tfid_bow_lemmatized.get_feature_names())
```

```
tfid_stemmed_df.head()
```

Создаём и обучаем Word2Vec модель

```
w2v = Word2Vec(size=200, window=5, min_count=2, sg=1, hs=0, negative=10, workers=32, seed=34)
```

```
w2v.build_vocab(df['tweet_token'], progress_per=100)
```

```
w2v.train(df['tweet_token'], total_examples=w2v.corpus_count, epochs=20)
```

Смотрим результат

```
w2v.wv.most_similar(positive="dinner")
```

```
w2v.wv.most_similar(positive="trump")
```

Создаём векторное представление твита с помощью суммы усреднённых значений векторов слов, входящих в твит.

Некоторых слов нет в словаре, поэтому делаем обязательную проверку

```
r = []
for tweet in df['tweet_token']:
  vec = np.zeros((1, 200))
  vs = [w2v.wv[word].reshape((1, 200)) for word in tweet if word in w2v.wv.vocab]

  for v in vs:
    vec += v

  r.append(vec/len(tweet))
```

```
r = np.array(r).reshape(49159, 200)
r.shape
```

```
w2v_df = pd.DataFrame(r)
```
